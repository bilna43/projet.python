```python
Poly_IRIS = gp.read_file('/Users/nabilboukir/Desktop/Projet_python/iris-2013-01-01/iris-2013-01-01.shx')

#Fichier contenant le contour géographique du découpade en IRIS et leur type. La projection est WGS84
#Données issues de https://www.data.gouv.fr/fr/datasets/contour-des-iris-insee-tout-en-un/
```

```python
Poly_IRIS = Poly_IRIS.sort_values("DCOMIRIS")
Poly_IRIS = Poly_IRIS.rename(columns = {'DCOMIRIS' : 'code_iris'})
Poly_IRIS.head()
```

```python
Poly_IRIS.head()
```

```python

Poly_IRIS['code_valide'] = Poly_IRIS['code_iris'].apply(lambda str: 'True' if str.isnumeric() else 'False')

Poly_Iris_inutilisable = Poly_IRIS.loc[Poly_IRIS['code_valide'] == 'False'] 
```

```python
L2 = Poly_Iris_inutilisable['NOM_COM'].drop_duplicates()
L2
```

```python
Poly_IRIS = Poly_IRIS.loc[Poly_IRIS['code_valide'] == 'True']
Poly_IRIS.head()
```

```python
Poly_IRIS['code_iris'] = pd.to_numeric(Poly_IRIS['code_iris'])

#essayer dans les deux bases de donnees d'utiliser apply(to_numeric, errors='ignore') pour voir si on peut garder la corse
```

```python
df2 = df.merge(Poly_IRIS, left_on = 'code_iris' , right_on = 'code_iris')
```

```python
df2['TYP_IRIS'].drop_duplicates()
```

```python
df2.keys()
#On va étudier la conse par IRIS donc, supprime les colonnes dont on a pas besoin, à savoir 'nom_commune', nom_iris' , 'segment_de_client', 'numero_de_voie', 'indice_de_repetition', 'type_de_voie'
# 'libelle_de_voie', 'code_commune', 'tri_des_adresses', 'adresse', 'code_valide_x', 'code_valide_y'
```

```python
df2 = df2.drop(columns=['nom_commune' , 'nom_iris' , 'segment_de_client' , 'numero_de_voie' , 'indice_de_repetition' , 'type_de_voie' , 'libelle_de_voie' , 'code_commune' , 'tri_des_adresses' , 'adresse' , 'code_valide_x' , 'code_valide_y'])
```

```python
#On aggrège la consommation et le nombre de logeùent par IRIS pour mener notre étude descriptive
df2[['code_iris' , 'consommation_annuelle_totale_de_l_adresse_mwh' , 'annee']].groupby(['code_iris','annee']).sum()
df2.keys()
```

```python
df2 = df2.merge(df2.groupby(['DEPCOM','annee']).consommation_annuelle_totale_de_l_adresse_mwh.sum(), left_on = ['DEPCOM' , 'annee'] , right_on = ['DEPCOM' , 'annee'] , suffixes = ('','_agreg' ))
df2.head()
```

```python
df2.groupby(['DEPCOM','annee']).consommation_annuelle_totale_de_l_adresse_mwh.sum()
#On peut vérifier qu'on a obtenu les valeurs souhaités
```

```python
df2 = df2.merge(df2.groupby(['DEPCOM','annee']).nombre_de_logements.sum(), left_on = ['DEPCOM' , 'annee'] , right_on = ['DEPCOM' , 'annee'] , suffixes = ('','_agreg' ))
```

```python
gdf = gp.GeoDataFrame(df2 ,geometry='geometry' , crs = 'WGS84')
```

```python
gdf['Departement'] = gdf['DEPCOM'].apply(lambda str: str[:2])
gdf.head()
```
